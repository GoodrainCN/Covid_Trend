```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
import numpy as np
####
data = pd.read_csv('OR.csv', usecols=[15]) #x
fish = pd.read_csv('OR.csv', usecols=[2]) #y
data = np.array(data) #x
fish = np.array(fish) #y
####
import matplotlib.pyplot as plt
plt.scatter(data, fish)
plt.show()
####
X_train, X_test, Y_train, Y_test = train_test_split(data, fish, train_size=.8)
plt.scatter(X_train,Y_train, color='b')
plt.scatter(X_test,Y_test, color='r')
plt.show()
####
model = LinearRegression()
model.fit(X_train, Y_train)
####
a = model.intercept_
b = model.coef_
print(a[0])
print(b[0][0])
####
model.score(X_test,Y_test)
output_y = model.predict(X_test)
plt.scatter(X_test,output_y, color='b')
plt.scatter(X_test,Y_test, color='r')
x = []
y = []
y1 = []
for i in range(len(data)):
    x.append(i)
    y.append((i*b[0][0])+a[0])
    y1.append(0)
#print(y)
#x = x.array(x)
#y = y.array(y)
plt.plot(x,y)
#plt.plot(x,y1)
```
