```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
FL_data = pd.read_csv('cleaned/FL.csv', usecols=[5])
FL_data = FL_data[225:350]
plt.plot(FL_data)
```

```python
x = np.arange(225,350,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
import numpy as np
####
data = np.array(x) #x
data = data.reshape(-1,1)
date = np.array(FL_data) #y
####
import matplotlib.pyplot as plt
plt.scatter(data, date)
plt.show()
####
X_train, X_test, Y_train, Y_test = train_test_split(data, date, train_size=.8)
plt.scatter(X_train,Y_train, color='b')
plt.title('FL2020winter_Original')
plt.savefig('result/FL2020winter_Original.jpg')

plt.scatter(X_test,Y_test, color='r')
plt.title('FL2020winter_Train')
plt.savefig('result/FL2020winter_Train.jpg')
plt.show()
####
model = LinearRegression()
model.fit(X_train, Y_train)
####
a = model.intercept_
b = model.coef_
fb = int(a[0])
fk = int(b[0][0])
print(f"b={fb}")
print(f"k={fk}")
if fb >= 0:
    print(f"f(x) = {fk}x + {fb}")
elif fb < 0:
    print(f"f(x) = {fk}x {fb}")
####
model.score(X_test,Y_test)
output_y = model.predict(X_test)
plt.scatter(X_test,output_y, color='b')
plt.scatter(X_test,Y_test, color='r')
x = []
y = []
y1 = []
for i in range(len(data)):
    x.append(i)
    y.append((i*b[0][0])+a[0])
    y1.append(0)
plt.plot(x,y)
plt.title('FL2020winter_Result')
plt.savefig('result/FL2020winter_Result.jpg')
```
